```python
# Create lists of NYC Hotel Name, latitude, longitude, rating, and review
nyc_hotel_name = []
nyc_hotel_category = []
nyc_hotel_lat = []
nyc_hotel_lon = []
nyc_hotel_rating = []
nyc_hotel_review_count = []
for i in range(len(nyc_hotels)):
    nyc_hotel_name.append(nyc_hotels[i]['name'])
    nyc_hotel_category.append(NYC_hotels['businesses'][i]['categories'][0]['title'])
    nyc_hotel_lat.append(nyc_hotels[i]['coordinates']['latitude'])
    nyc_hotel_lon.append(nyc_hotels[i]['coordinates']['longitude'])
    nyc_hotel_rating.append(nyc_hotels[i]['rating'])
    nyc_hotel_review_count.append(nyc_hotels[i]['review_count'])   
print(nyc_hotel_review_count)
```
