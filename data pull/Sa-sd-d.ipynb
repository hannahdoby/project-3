```python
import json
import pprint
import requests
import er_config as er_key

url = "https://api.yelp.com/v3/businesses/search?location="
headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {er_key}"
}

locations= ["SanAntonio","SanDiego", "Dallas"]
interests = ["hotels","food"]

data= {}

# Loop through the locations and interests to pull the data from the Yelp API
for location in locations:
    i = 0
    term = interests[i]
    response = requests.get(f"{url}{location}&term={term}&sort_by=best_match&limit=40", headers=headers)
    locals()[f"{location}_{term}"] = json.loads(response.text)
    i += 1
    term = interests[i]
    response = requests.get(f"{url}{location}&term={term}&sort_by=best_match&limit=40", headers=headers)
    locals()[f"{location}_{term}"] = json.loads(response.text)

#Define locals variables
SanAntonio_food = locals()["SanAntonio_food"]
SanAntonio_hotels = locals()["SanAntonio_hotels"]
SanDiego_food = locals()["SanDiego_food"]
SanDiego_hotels = locals()["SanDiego_hotels"]
Dallas_food = locals()["Dallas_food"]
Dallas_hotels = locals()["Dallas_hotels"]


#Pretty print varables
pprint.pprint(SanAntonio_food)
pprint.pprint(SanAntonio_hotels)
pprint.pprint(SanDiego_food)
pprint.pprint(SanDiego_hotels)
pprint.pprint(Dallas_food)
pprint.pprint(Dallas_hotels)
```

```python
import os

#Make file path for JSON Foler 
folder_name = 'JSON'

# Construct the file path for SA FOOD
json_file_path = os.path.join(folder_name, 'SanAntonio_food.json')

# Export to JSON
with open(json_file_path, 'w') as outfile:
    json.dump(SanAntonio_food, outfile)


# Construct the file path for SA HOTELS
json_file_path = os.path.join(folder_name, 'SanAntonio_hotels.json')

# Export to JSON
with open(json_file_path, 'w') as outfile:
    json.dump(SanAntonio_hotels, outfile)


# Construct the file path SD FOOD
json_file_path = os.path.join(folder_name, 'SanDiego_food.json')

# Export to JSON
with open(json_file_path, 'w') as outfile:
    json.dump(SanDiego_food, outfile)

# Construct the file path SD HOTELS
json_file_path = os.path.join(folder_name, 'SanDiego_hotels.json')

# Export to JSON
with open(json_file_path, 'w') as outfile:
    json.dump(SanDiego_hotels, outfile)

# Construct the file path DALLAS FOOD
json_file_path = os.path.join(folder_name, 'Dallas_food.json')

# Export to JSON
with open(json_file_path, 'w') as outfile:
    json.dump(Dallas_food, outfile)

# Construct the file path DALLAS HOTELS
json_file_path = os.path.join(folder_name, 'Dallas_hotels.json')

# Export to JSON
with open(json_file_path, 'w') as outfile:
    json.dump(Dallas_hotels, outfile)

```

```python
# Load JSON data from the file
with open('JSON/SanAntonio_food.json', 'r') as json_file:
    data = json.load(json_file)

# Print the loaded data to check its structure
print(json.dumps(data, indent=4))

# Attempt to access the 'businesses' key
try:
    businesses = data['businesses']
    for business in businesses:
        print(business['name'])
except KeyError:
    print("The 'businesses' key does not exist in the loaded data.")
```

```python
import csv


#Define extracted data     
extracted_data_SA_food = []
for business in data['businesses']:
    name = business['name']
    categories = ', '.join([category['title'] for category in business['categories']])
    review = business['rating']
    review_count = business['review_count']
    
    # Append the extracted data as a dictionary
    extracted_data_SA_food.append({
        'Name': name,
        'Category': categories,
        'Review': review,
        'Review Count': review_count
    })


# Define the CSV filename
SA_Food_csv_filename = 'extracted_data_SA_food.csv'

# absolute path to the current directory
current_directory = os.getcwd()

# Define the CSV filename and folder
folder_path = os.path.join(current_directory, 'CSV')
os.makedirs(folder_path, exist_ok=True)
SA_Food_csv_filename = os.path.join(folder_path, 'extracted_data_SA_food.csv')

# Write the extracted data to a CSV file
with open(SA_Food_csv_filename, 'w', newline='') as csv_file:
    fieldnames = ['Name', 'Category', 'Review', 'Review Count']
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    # Write the header
    csv_writer.writeheader()

    # Write the extracted data
    csv_writer.writerows(extracted_data_SA_food)
```
